```python
a = list (range(1, 1001))
b = list (range(1, 1001))
c = list (range(1, 1001))
experimental_result = [None]*1000
theorectical_result = [None]*1000
constant = [None]*1000

print(a)
print(b)
print(c)
print(experimental_result)
print(theorectical_result)

size = 1000
Sum=0
average_constant=0
```

```python
from time import perf_counter_ns

for n in range(1,size+1):
  start_time = perf_counter_ns()
  for i in range(1,n):
    for j in range(i,n):
      for k in range(j*j,n):
        Sum += a[i]*b[j]*c[k]

  end_time = perf_counter_ns()
  experimental_result[n-1] = (end_time - start_time)/10**12
  print("Elapsed time (s):", experimental_result[n-1])
  print("Number of Elements (n):", n)
```

```python
from math import sqrt

for n in range(1,size+1):
  theorectical_result[n-1] = (n*n*sqrt(n))
  print("Elapsed time (s):", theorectical_result[n-1])
  print("Number of Elements (n):", n)
```

```python
for i in range(0,size):
  constant[i]=theorectical_result[i]/experimental_result[i]

for i in range(0,size):
  average_constant=average_constant+constant[i]
  experimental_result[i]=constant[i]*experimental_result[i]

average_constant=average_constant/1000
print(average_constant)
print(constant)
```

```python
print(constant[199])
print(constant[399])
print(constant[599])
print(constant[799])
print(constant[999])
```
