```python
import random
import numpy as np
from time import perf_counter_ns
```

```python
def costofNetwork(clients, position, start, end):
    cost = 0
    for i in range(start, end + 1):
        cost += abs(position - i)*clients[i]

    return cost


def findMinCost(clients, start, end):
    minIndex = 0
    min = 1000
    for i in range(start, end + 1):
        Cost = costofNetwork(clients, i, start, end)
        if (Cost < min):
            min = Cost
            minIndex = i

    return min


def clientsgen(n):
  temp = [None]*n;
  for i in range(n):
    temp[i] = random.randint(1,10)
  
  return temp
```

```python
number = 101
experimental_result = [None]*number
theorectical_result = [None]*number
numbers = [None]*number
constant = [None]*number
experimental_result_corrected = [None]*number
```

```python
for i in range(0,number):
    theorectical_result[i] = (i)*(i)
```

```python
for cnum in range(8,number+8):
  
  clients = clientsgen(cnum)
  n = len(clients)
  serverCount = 3
  dp = np.zeros((n,n,serverCount))

  for i in range(0,n):
    for j in range(0,n):
      for k in range(0,serverCount):
          dp[i][j][k] = 2342342523

  start_time = perf_counter_ns()
  for j in range(1,n):
      dp[0][j][0] = findMinCost(clients, 0,j)
      dp[j][n-1][0] = findMinCost(clients,j+1,n-1)


  for i in range(0,n):
    for j in range(0,n):
      for k in range(0,serverCount):
        if(j-i <= k):
          dp[i][j][k] = 0


  for j in range(2,n):
    for i in range(0,j):
        for k in range(1,serverCount):
          dp[i][j][k] = min(dp[0][j][k], dp[0][i][k-1] + dp[i+1][j][0])

  end_time = perf_counter_ns()
  avg_time = (end_time - start_time)/10**6
  experimental_result[cnum-8] = avg_time/1000
  numbers[cnum-8] = n
  
```

```python
import matplotlib.pyplot as plt
  
# line 1 points
x1 = numbers
y1 = experimental_result
# plotting the line 1 points 
plt.plot(x1, y1, label = "Experimental results")
  
# naming the x axis
plt.xlabel('Number of elements(n)')
# naming the y axis
plt.ylabel('Time(t in microseconds)')
# giving a title to my graph
plt.title('Experimental results ')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```

```python
average_constant = 0
for i in range(0,number):
  constant[i]=theorectical_result[i]/experimental_result[i]

for i in range(0,number):
  average_constant=average_constant+constant[i]

average_constant=average_constant/number

for i in range(0,number):
  experimental_result_corrected[i]=average_constant*experimental_result[i]

print(average_constant)
print(constant)
```
